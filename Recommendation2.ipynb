# Recommendation System

# Libraries

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import TruncatedSVD
```

# Dataset

### First Dataset

```python
ratings = pd.read_csv('ratings.csv')
columns = ['book_id', 'user_id', 'rating']
```

### Second Dataset

```python
books = pd.read_csv('books.csv')
coumns = ['id', 'book_id', 'best_book_id', 'work_id', 'books_count',
'isbn', 'isbn13', 'authors', 'original_publication_year', 'original_title', 'title',
'language_code', 'average_rating', 'ratings_count', 'work_ratings_count', 'work_text_reviews_count',
'ratings_1', 'ratings_2', 'ratings_3', 'ratings_4', 'ratings_5', 'image_url', 'small_image_url']
```

### Mixed Data

```python
combined_data = pd.merge(ratings, books, on='book_id')
combined_data.head()
```

# Main Program

```python
rating_utility = combined_data.pivot_table(values = 'rating', index = 'user_id', columns = 'title', fill_value=0)
rating_utility.head()
```

```python
X = rating_utility.T
```

```python
SVD = TruncatedSVD(n_components=30)
transposed_matrix = SVD.fit_transform(X)
transposed_matrix.shape
```

```python
corr_matrix = np.corrcoef(transposed_matrix)
books = rating_utility.columns
book_list = list(books)
book_1984 = book_list.index('1984')
book_1984
```

```python
corr_1984 = corr_matrix[4]
```

```python
list(books[(corr_1984 < 1.0) & (corr_1984 > 0.8)])
```

```python
list(books[(corr_1984 < 1.0) & (corr_1984 > 0.7)])
```
