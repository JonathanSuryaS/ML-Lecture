# Libraries

```python
from sklearn import tree
import pandas as pd
import pydotplus
from IPython.display import Image
```

# Dataset

```python
data = pd.read_csv('PlayGolf.csv')
data
```

Convert categorical variable into dummy / indicator variable or (binary value)

```python
one_hot_data = pd.get_dummies(data[['Outlook', 'Temperature', 'Humidity', 'Windy']])
one_hot_data
```

# Modeling

```python
clf = tree.DecisionTreeClassifier()
clf_train = clf.fit(one_hot_data, data['Play'])
```

```python
print(tree.export_graphviz(clf_train, None))
dot_data = tree.export_graphviz(clf_train, out_file=None, feature_names=list(one_hot_data.columns.values),class_names=['Not_Play', 'Play'], rounded=True, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png('playgolf.png')
Image(graph.create_png())
```
