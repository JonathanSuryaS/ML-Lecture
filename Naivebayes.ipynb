# Hands on Naive Bayes

# Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Dataset

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/joh4nnes/MLData/main/Social_Network_Ads.csv')
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Splitting the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Model

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(classifier.predict(sc.transform([[30, 87000]])))
```

# Evaluation

```python
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train, y_train, clf = classifier)
plt.title('Training set')
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_test, y_pred, clf = classifier)
plt.title('Predicting test set')
plt.show()
```
